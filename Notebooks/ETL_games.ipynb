```python
# Este fue el codigo correcto para desanidar la informacion (pasar de JSon a CSV).
import json
import pandas as pd

# Inicializa una lista para almacenar los objetos JSON
lista_datos = []

# Abre el archivo JSON en modo de lectura
with open("Jsons/output_steam_games.json", 'r') as archivo_json:
    # Itera sobre las líneas del archivo y carga cada línea como un objeto JSON
    for linea in archivo_json:
        objeto_json = json.loads(linea)
        lista_datos.append(objeto_json)

# Convierte la lista en un DataFrame
df_games = pd.DataFrame(lista_datos)

# Guarda el DataFrame en un archivo CSV
df_games.to_csv("CSV/output_steam_prueba.csv", index=False)  # El argumento 'index=False' evita que se escriba el índice en el archivo CSV
```

```python
# Para encontrar todas las fechas de lanzamiento que sean como JUN 2009.

import pandas as pd

# Lee el archivo CSV en un DataFrame
df_games = pd.read_csv('CSV/output_steam_games_limpio.csv')

# Define una función para identificar las fechas que no cumplen con el formato
def is_invalid_date(date_string):
    try:
        pd.to_datetime(date_string, format='%Y-%m-%d')
        return False  # La fecha cumple con el formato
    except ValueError:
        return True   # La fecha no cumple con el formato

# Aplica la función para identificar fechas inválidas
fechas_invalidas = df_games['release_date'].apply(is_invalid_date)

# Muestra las filas con fechas inválidas
print(df_games[fechas_invalidas])
```

```python
# para crear la columna release_year y extraer el año.

import pandas as pd

# Lee el archivo CSV original
df_games = pd.read_csv('CSV/output_steam_games_limpio.csv')

# Extrae el año de la columna 'release_date' y crea la nueva columna 'release_year'
df_games['release_year'] = pd.to_datetime(df_games['release_date'], format='%d/%m/%Y', errors='coerce').dt.year

# Reemplaza los valores no finitos (NaN o inf) en 'release_year' con un valor predeterminado, por ejemplo, 0
df_games['release_year'].fillna(0, inplace=True)

# Convierte la columna 'release_year' a entero
df_games['release_year'] = df_games['release_year'].astype(int)

# Reescribe el archivo CSV con la nueva columna
df_games.to_csv('CSV/output_steam_games_limpio.csv', index=False)
```
